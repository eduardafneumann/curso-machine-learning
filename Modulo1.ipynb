```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('./Anexo_pinguins-2.csv')
```

```python
sns.heatmap(df.isnull())
```

```python
sns.barplot(df.groupby(by=['Ilha']).size())
```

```python
sns.barplot(df.groupby(by=['Espece']).size())
plt.xlabel('Espécie')
plt.show()
```

```python
sns.pairplot(data=df.drop(['sexo', 'Ilha'], axis=1), hue="Espece")
plt.show()
```

```python
sns.pairplot(data=df.drop(['Ilha'], axis=1).loc[df['Espece'] == 'Adelie'], hue="sexo")
plt.show()
```

```python
sns.pairplot(data=df.drop(['Ilha'], axis=1).loc[df['Espece'] == 'Chinstrap'], hue="sexo")
plt.show()
```
